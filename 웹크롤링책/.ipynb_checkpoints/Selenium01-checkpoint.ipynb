```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get("https://www.livesport.com/kr/team/manchester-united/ppjDR086/")
```

```python
from bs4 import BeautifulSoup
```

```python
page = driver.page_source
```

```python
page
```

```python
bs_obj = BeautifulSoup(page, "html.parser")
```

```python
bs_obj
```

```python
win = bs_obj.find_all("span", class_="wld wld--w")
```

```python
win
```

```python
draw = bs_obj.find_all("span", class_="wld wld--d")
```

```python
draw
```

```python
lose = bs_obj.find_all("span", class_="wld wld--l")
```

```python
lose
```

```python
#만약 연장이후에 승이거나 승부차기 승이라면 다른 클래스가 있을수 있으므로 확인은 꼭 해야한다
n_win = len(win)
n_draw = len(draw)
n_lose = len(lose)
```

```python
print(n_win)
print(n_draw)
print(n_lose)
```

```python
report = {'win': n_win, "draw" : n_draw, "lose" : n_lose}
```

```python
report
```

```python
max_n = max(report.values())
report.keys()
```

```python
# max_n=5 이므로 key값은 win

for key in report:
    if(report[key]==max_n):
        print(key)
```
