```python
# 사이트에 자료 요청
import urllib.request
# json 데이터 핸들링
import json
# DataFrame 자료형 활용
import pandas as pd
# json 데이터를 pandasDataFrame 변환
from pandas.io.json import json_normalize
```

```python
r = urllib.request.urlopen(url)
```

```python
json_raw = r.read()
```

```python
json_raw
```

```python
json_utf8 = json_raw.decode("utf-8")
```

```python
json_utf8
```

```python
type(json_utf8)
```

```python
# str타입을 딕셔너리 타입으로 변경
json_complete = json.loads(json_utf8)
```

```python
type(json_complete)
```

```python
pd.json_normalize(json_complete)
```

```python
box_result = pd.json_normalize(json_complete['landBizInfo']['row'])
```
