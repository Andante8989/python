```python
# 크롤링 라이브러리 임포트
from bs4 import BeautifulSoup # 웹에서 가져온 데이터를 정제하는 역할
import requests # 파이썬에서 다이렉트로 자료를 요청하는 역할
from selenium import webdriver # 크롬창을 켜주는 역할
import time # 코드 중간에 딜레이를 넣는 역할
from selenium.webdriver.common.by import By
import codecs
```

```python
# 브라우저 켜기
browser = webdriver.Chrome('chromedriver.exe')

# cgv 무비차트 페이지 접속
browser.get('http://www.cgv.co.kr/movies/?lt=1&ft=0')

time.sleep(2)

browser.find_element(By.XPATH,'//*[@id="contents"]/div[1]/div[3]/button/i').click()
```

```python
# 소스코드 저장
source = browser.page_source

# 저장된 소스코드 파싱
parsed_source = BeautifulSoup(source, "html.parser")
```

```python
parsed_source
```

```python
# 전체 소스코드를 가져왔으므로 필요한정보를 포함하는 가장작은 단위를 찾아 가져옴
div_source = parsed_source.find_all("div", id="contents")
```

```python
len(div_source)
```

```python
div_source_list = div_source[0]
```

```python
div_title = div_source_list.find_all("strong", class_="title")
```
