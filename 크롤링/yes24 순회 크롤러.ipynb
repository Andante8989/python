```python
# 크롤링 라이브러리 임포트
from bs4 import BeautifulSoup # 웹에서 가져온 데이터를 정제하는 역할
import requests # 파이썬에서 다이렉트로 자료를 요청하는 역할
from selenium import webdriver # 크롬창을 켜주는 역할
import time # 코드 중간에 딜레이를 넣는 역할
from selenium.webdriver.common.by import By
import codecs
```

```python
# 웹드라이버 실행
browser = webdriver.Chrome('chromedriver.exe')

# yes24 베스트 첫페이지로 이동
browser.get('http://www.yes24.com/24/category/bestseller?CategoryNumber=001&sumgb=06&fetchSize=40&PageNumber=1')

# 소스코드 저장
source = browser.page_source

# 저장된 소스코드 파싱
parsed_source = BeautifulSoup(source, "html.parser")

table_tag_title_list = parsed_source.find_all("table", class_="list")
len(table_tag_title_list)
```

```python
# 책 제목 가져옴
title_list = parsed_source.find_all("td", class_="goodsTxtInfo")
for i in title_list :
    title = i.find_all("a")
    print(title[0].text)
    title_list2.append(title[0].text)
    
```

```python
# 책 저자 가져옴
author_list = parsed_source.find_all("div", class_="aupu")
for i in author_list :
    print(i.text.replace("\n","").replace("\t","").replace(" ",""))
    author_list2.append(i.text.replace("\n","").replace("\t","").replace(" ",""))
```

```python
# 책 가격 가져옴
price_list = parsed_source.find_all("span", class_="priceB")
for i in price_list :
    print(i.text)
    price_list2.append(i.text)
```

```python
browser = webdriver.Chrome('chromedriver.exe')


f = codecs.open("c:/crawler2/yes24.txt", mode="w", encoding="utf-8")

book_rank = 0

for a in range(1,26):
    
    browser.get('http://www.yes24.com/24/category/bestseller?CategoryNumber=001&sumgb=06&fetchSize=40&PageNumber=%s' % a)
    time.sleep(3)
    
    title_list2 = []
    price_list2 = []
    author_list2 = []
    
    # 소스코드 저장
    source = browser.page_source

    # 저장된 소스코드 파싱
    parsed_source = BeautifulSoup(source, "html.parser")

    table_tag_title_list = parsed_source.find_all("table", class_="list")
    
    title_list = parsed_source.find_all("td", class_="goodsTxtInfo")
    for i in title_list :
        title = i.find_all("a")
        title_list2.append(title[0].text)
        
        # 책 저자 가져옴
    author_list = parsed_source.find_all("div", class_="aupu")
    for i in author_list :
        author_list2.append(i.text.replace("\n","").replace("\t","").replace(" ",""))
        
    # 책 가격 가져옴
    price_list = parsed_source.find_all("span", class_="priceB")
    for i in price_list :
        price_list2.append(i.text)
        
    for i in range(len(title_list2)):
        book_rank += 1
        f.write("%s등" % book_rank + "\n")
        f.write(title_list2[i] + "\n")
        f.write(author_list2[i] + "\n")
        f.write(price_list2[i] + "\n")

f.close()
```
