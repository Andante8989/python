```python
# 크롤링 라이브러리 임포트
from bs4 import BeautifulSoup # 웹에서 가져온 데이터를 정제하는 역할
import requests # 파이썬에서 다이렉트로 자료를 요청하는 역할
from selenium import webdriver # 크롬창을 켜주는 역할
import time # 코드 중간에 딜레이를 넣는 역할
from selenium.webdriver.common.by import By
import codecs
```

```python
# 웹드라이버 실행
browser = webdriver.Chrome('chromedriver.exe')

# 교보문고 베스트 첫페이지로 이동
browser.get('http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf')

# 소스코드 저장
source = browser.page_source

# 저장된 소스코드 파싱
parsed_source = BeautifulSoup(source, "html.parser")

# 위에는 페이지 전체를 가져온것이므로 부분으로 가져와야됨, 정보를 포함하는 최소단위 ul태그 가져옴
div_tag_title_list = parsed_source.find_all("ul", class_="list_type01")
len(div_tag_title_list)  # 1개라고 출력됨

# 리스트여서 1개라고 나오므로 인덱싱으로 꺼내서 추출해야됨
ul_tag_type01 = div_tag_title_list[0]

len(div_tag_title_list)
```

```python
# 책제목 50개 가져옴, 리스트에저장
div_tag_title_list = ul_tag_type01.find_all("div", class_="title")
for item in div_tag_title_list:
    print(item.text.strip())
    title_list.append(item.text.strip())
```

```python
# 저자 50 개 가져옴, 리스트에저장
div_tag_writer_list = ul_tag_type01.find_all("div", class_="author")
for item in div_tag_writer_list:
    print(item.text.strip().replace("\n","").replace("\t",""))
    author_list.append(item.text.strip().replace("\n","").replace("\t",""))
```

```python
# 가격 50 개가져옴, 리스트에저장
div_tag_price_list = ul_tag_type01.find_all("strong", class_="book_price")
for item in div_tag_price_list:
    print(item.text)
    price_list.append(item.text)
```

```python
browser = webdriver.Chrome('chromedriver.exe')


f = codecs.open("c:/crawler2/kyobo.txt", mode="w", encoding="utf-8")

book_rank = 0

for i in range(1, 5):
   
    browser.get('https://www.kyobobook.co.kr/bestSellerNew/bestseller.laf?targetPage=%s' % i)
    time.sleep(3)
    
    title_list = []
    price_list = []
    author_list = []
    
    # 소스코드 저장
    source = browser.page_source

    # 저장된 소스코드 파싱
    parsed_source = BeautifulSoup(source, "html.parser")

    # 위에는 페이지 전체를 가져온것이므로 부분으로 가져와야됨, 정보를 포함하는 최소단위 ul태그 가져옴
    div_tag_title_list = parsed_source.find_all("ul", class_="list_type01")
    len(div_tag_title_list)  # 1개라고 출력됨

    # 리스트여서 1개라고 나오므로 인덱싱으로 꺼내서 추출해야됨
    ul_tag_type01 = div_tag_title_list[0]
    
    # 제목 50 개 가져옴, 리스트에 저장
    div_tag_title_list = ul_tag_type01.find_all("div", class_="title")
    for item in div_tag_title_list:
        title_list.append(item.text.strip())
        
    # 저자 50 개 가져옴, 리스트에저장
    div_tag_writer_list = ul_tag_type01.find_all("div", class_="author")
    for item in div_tag_writer_list:
        author_list.append(item.text.strip().replace("\n", "").replace("\t", ""))
        
    # 가격 50 개가져옴, 리스트에저장
    div_tag_price_list = ul_tag_type01.find_all("strong", class_="book_price")
    for item in div_tag_price_list:
        price_list.append(item.text)
    
    for i in range(len(title_list)):
        book_rank += 1
        f.write("%s등" % book_rank + "\n")
        f.write(title_list[i] + "\n")
        f.write(author_list[i] + "\n")
        f.write(price_list[i] + "\n")
       
    
f.close()
```
